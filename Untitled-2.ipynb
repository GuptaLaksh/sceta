```python
def find_inflection_points(x, y, smooth_sigma=1.0, min_distance=None):
    """
    Find inflection points of a curve given x and y coordinates.
    
    Inflection points are where the second derivative (curvature) changes sign.
    
    Parameters:
    -----------
    x : array-like
        X coordinates (will be sorted internally)
    y : array-like
        Y coordinates
    smooth_sigma : float
        Gaussian smoothing parameter (larger = more smoothing).
        Use 0.5-2.0 for noisy data.
    min_distance : int or None
        Minimum distance between inflection points (in indices).
        Helps avoid duplicate detections in noisy regions.
        
    Returns:
    --------
    x_inflect : array
        X coordinates of inflection points
    y_inflect : array
        Y coordinates of inflection points
    curvature : array
        Second derivative (curvature) at inflection points
        
    Example:
    --------
    >>> x = np.linspace(-3, 3, 200)
    >>> y = x**3 - 2*x**2 + x + 1
    >>> x_inf, y_inf, curv = find_inflection_points(x, y, smooth_sigma=0.5)
    """
    from scipy.ndimage import gaussian_filter1d
    
    # Convert to arrays and sort by x
    x = np.asarray(x)
    y = np.asarray(y)
    sort_idx = np.argsort(x)
    x_sorted = x[sort_idx]
    y_sorted = y[sort_idx]
    
    # Smooth to reduce noise
    y_smooth = gaussian_filter1d(y_sorted, sigma=smooth_sigma)
    
    # Compute first derivative
    dy = np.gradient(y_smooth, x_sorted)
    
    # Compute second derivative (curvature)
    d2y = np.gradient(dy, x_sorted)
    
    # Find sign changes in second derivative
    sign_changes = np.where(np.diff(np.sign(d2y)))[0]
    
    # Filter by minimum distance if specified
    if min_distance is not None and len(sign_changes) > 1:
        filtered = [sign_changes[0]]
        for idx in sign_changes[1:]:
            if idx - filtered[-1] >= min_distance:
                filtered.append(idx)
        sign_changes = np.array(filtered)
    
    x_inflect = x_sorted[sign_changes]
    y_inflect = y_sorted[sign_changes]
    curvature = d2y[sign_changes]
    
    return x_inflect, y_inflect, curvature
```

```python
def find_rgb_inflection(df, color_col, mag_col):
    # Sort by color
    df_sorted = df.sort_values(color_col)
    
    colors = df_sorted[color_col].values
    mags = df_sorted[mag_col].values
    
    # Smooth with rolling mean first
    from scipy.ndimage import gaussian_filter1d
    mags_smooth = gaussian_filter1d(mags, sigma=2)
    
    # Second derivative
    d2 = np.gradient(np.gradient(mags_smooth))
    
    # Find sign changes
    sign_changes = np.where(np.diff(np.sign(d2)))[0]
    
    return colors[sign_changes], mags[sign_changes]

# to select stars with pm > 80.
df = func_analysis.cleandata(dfclusters[clustername], 80) 

#  to remove stars with dF275W < 1 and dF336W < 1
# df = dfclusters[clustername][~((dfclusters[clustername]['dF275W'] > 1) & (dfclusters[clustername]['dF336W'] > 1))]

plt.tight_layout()
plt.rcParams['figure.dpi']=100
fig=plt.figure(figsize=(10,10))

plt.scatter(dfclusters[clustername]['F606W'] - dfclusters[clustername]['F814W'], dfclusters[clustername]['F606W'], 
            marker = 'o', s=2, alpha=1, color='black', edgecolors='none')

c_isohold = models['iso131'][str(ageiso)]
canonicaliso_dict = func_analysis.corrections(aDM+0.01, reddening-0.005, {'main': c_isohold[0:1300]})
canonicaliso = canonicaliso_dict['main']  
canonicaliso['F606W'] -= 0.1

x_inf, y_inf, curv = find_inflection_points(x, y, smooth_sigma=0.5)


plt.plot(canonicaliso['F606W'] - canonicaliso['F814W'],
            canonicaliso['F606W'],
            color=names['colors'][0], lw=1, linestyle='-', markersize=0.1,
            marker='o', label='He-canonical Iso')

plt.ylim(10, 35)
plt.xlim(-2, 6)
plt.gca().invert_yaxis()
plt.legend()

plt.show()  
```
